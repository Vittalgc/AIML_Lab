```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
```

```python
data = pd.read_csv("weight-height.csv")
data.describe()
```

```python
X = data[["Height"]]
y = data["Weight"]
```

```python
x_train, x_test, y_train, y_test = tts(X, y, test_size=0.25, random_state = 1)
```

```python
x_train.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data["Height"], y=y)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
pd.DataFrame({"True Values" : y_test.head(), "Predicted Values" : y_pred[:5]})
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=data["Height"], y=y, line_kws = {"color": "r"})
```

### Mean Absolute Error

```python
mae(y_pred, y_test)
```

### Mean Squared Error

```python
mse(y_pred, y_test)
```
